```python
missing = pd.concat([df.isnull().sum(), 100 * df.isnull().mean()], axis=1)
missing.columns=['count', '%']
missing = missing.sort_values(by = ['count','%'], ascending = False)
missing
```

```python
# Removing Product Description and Zip Code due to high missing values

df.drop(columns = ['Product Description','Order Zipcode'], inplace = True)
```

```python
# Removing Customer info like name as it does not relate to the data anlaysis
customer_info = ['Customer Email','Customer Fname','Customer Lname', 'Customer Password','shipping date (DateOrders)','order date (DateOrders)']
df.drop(columns = customer_info, inplace = True)
df.info()
```

```python
product_info = ['Product Image', 'Product Card Id']
df.drop(columns = product_info, inplace = True)
df.info()
```
